```python
import numpy as np
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings ('ignore')
X = np.array ([[1],[2],[3]])
Y = np.array ([10,20,30])
model=LinearRegression()
model.fit (X,Y)
New_Question = np.array ([[7]])
Prediction = model.predict (New_Question)
print("My fast ml mode Predetion")
print(f"Possible 7th room rent:{Prediction[0]:.0f} A Thousand")
```
